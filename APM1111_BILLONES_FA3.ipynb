```python
#First Quartile
q1 = np.percentile(data, 25)
#Second Quartile
q2 = np.percentile(data, 50)
#Third Quartile
q3 = np.percentile(data, 75)
#9th Decile
d9 = np.percentile(data, 90)
#95th Decile
p95 = np.percentile(data, 95)

#valid count
val = len(data)
#mode
mode = stats.mode(data, keepdims=True).mode[0]
#median
median = np.median(data)
#mean
mean = np.mean(data)
#standard deviation
std_dev = np.std(data)
#variance
variance = np.var(data)
#skewness
skewness = stats.skew(data)
#standard error of skewness
std_err_skewness = skewness / np.sqrt(val)
#kurtosis
kurtosis = stats.kurtosis(data)
#standard error of kurtosis
std_err_kurtosis = kurtosis / np.sqrt(val)

#minimum and maximum
minimum = np.min(data)
maximum = np.max(data)
```
